```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
from datetime import datetime
%matplotlib inline
```

```python
df = pd.read_csv('datafest2018.csv')
```

```python
pd.set_option("display.max_columns",999)
```

```python
df.head()
```

```python
df['date'] = df['date'].apply(lambda x:dt.datetime.strptime(x,"%Y-%m-%d"))
```

```python
temp = df[(df['industry']=='HEALTH_CARE') & (df['country']=='US')]
```

```python
temp.reset_index(inplace=True)
temp1 = temp.drop('index',axis =1)
```

```python
temp1.head()
```

```python
temp3 = temp1[['jobId','date']]
```

```python
temp4 = temp3.groupby('jobId')
    
```

```python
mx = temp4['date'].max()
mn = temp4['date'].min()
fin = pd.concat([mn,mx],axis=1)
```

```python
fin1 = pd.DataFrame(fin)
```

```python
fin1.reset_index(inplace=True)
final = fin1.drop('index',axis=1)
```

```python
final.to_csv('Cleaned data_US health care.csv')
```
